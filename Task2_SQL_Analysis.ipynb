```python
import sqlite3
import pandas as pd
df = pd.read_csv("data.csv", encoding="latin1")
df.head()
```

```python
conn = sqlite3.connect("ecommerce.db")
df.to_sql("ecommerce", conn, if_exists="replace", index=False)
```

```python
pd.read_sql_query("SELECT * FROM ecommerce LIMIT 10;", conn)
```

```python
pd.read_sql_query("SELECT COUNT(*) AS total_rows FROM ecommerce;", conn)
```

```python
pd.read_sql_query("""
SELECT COUNT(*) AS missing_customer_ids
FROM ecommerce
WHERE CustomerID IS NULL;
""", conn)
```

```python
pd.read_sql_query("""
SELECT COUNT(DISTINCT CustomerID) AS unique_customers
FROM ecommerce;
""", conn)
```

```python
pd.read_sql_query("""
SELECT COUNT(DISTINCT StockCode) AS unique_products
FROM ecommerce;
""", conn)
```

```python
pd.read_sql_query("""
SELECT Description, SUM(Quantity) AS total_quantity
FROM ecommerce
GROUP BY Description
ORDER BY total_quantity DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql_query("""
SELECT SUM(Quantity * UnitPrice) AS total_revenue
FROM ecommerce;
""", conn)
```

```python
pd.read_sql_query("""
SELECT Country, SUM(Quantity * UnitPrice) AS revenue
FROM ecommerce
GROUP BY Country
ORDER BY revenue DESC;
""", conn)
```

```python
pd.read_sql_query("""
SELECT DATE(InvoiceDate) AS date,
       SUM(Quantity * UnitPrice) AS revenue
FROM ecommerce
GROUP BY DATE(InvoiceDate)
ORDER BY date;
""", conn)
```

```python
pd.read_sql_query("""
SELECT *
FROM ecommerce
WHERE Quantity < 0;
""", conn)
```

```python
pd.read_sql_query("""
SELECT CustomerID, SUM(Quantity * UnitPrice) AS total_spent
FROM ecommerce
WHERE CustomerID IS NOT NULL
GROUP BY CustomerID
ORDER BY total_spent DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql_query("""
SELECT AVG(Quantity * UnitPrice) AS avg_order_value
FROM ecommerce;
""", conn)
```

```python
pd.read_sql_query("""
SELECT CustomerID,
       SUM(Quantity) AS total_units,
       SUM(Quantity * UnitPrice) AS total_spent
FROM ecommerce
WHERE CustomerID IS NOT NULL
GROUP BY CustomerID
ORDER BY total_spent DESC;
""", conn)
```

```python
pd.read_sql_query("""
SELECT Country, COUNT(DISTINCT InvoiceNo) AS total_orders
FROM ecommerce
GROUP BY Country
ORDER BY total_orders DESC
LIMIT 5;
""", conn)
```

```python
pd.read_sql_query("""
SELECT Description, SUM(Quantity) AS total_returns
FROM ecommerce
WHERE Quantity < 0
GROUP BY Description
ORDER BY total_returns;
""", conn)
```
